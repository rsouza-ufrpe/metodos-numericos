# Simulação de Crescimento Populacional
Este caderno compara o **Modelo Maltusiano** (Crescimento Exponencial) com o **Modelo Logístico** (Crescimento com Limite de Recursos).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
P0 = 10     # População inicial
r = 0.1     # Taxa de crescimento
K = 500     # Capacidade de carga
t_max = 100 # Tempo total
dt = 0.1    # Passo de tempo

t = np.arange(0, t_max, dt)
P_malthus = np.zeros(len(t))
P_logistic = np.zeros(len(t))

P_malthus[0] = P0
P_logistic[0] = P0

# Implementação do Método de Euler
for i in range(1, len(t)):
    P_malthus[i] = P_malthus[i-1] + (r * P_malthus[i-1]) * dt
    P_logistic[i] = P_logistic[i-1] + (r * P_logistic[i-1] * (1 - P_logistic[i-1] / K)) * dt

# Plotagem
plt.figure(figsize=(10, 5))
plt.plot(t, P_malthus, label='Maltusiano')
plt.plot(t, P_logistic, label='Logístico')
plt.axhline(y=K, color='r', linestyle='--', label='Capacidade de Carga (K)')
plt.ylim(0, K + 100)
plt.title('Malthus vs Logístico')
plt.legend()
plt.show()
```
