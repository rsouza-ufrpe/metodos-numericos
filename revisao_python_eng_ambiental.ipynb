### Passo 2 — Coletar dados da API OpenAQ

A célula abaixo busca as últimas 100 medições de PM2.5 registradas no Brasil via API pública do OpenAQ.

> ⚠️ **Nota:** a API do OpenAQ pode exigir registro gratuito para chaves de acesso em versões mais recentes. Se a requisição falhar, os dados de fallback (segunda célula) serão usados.

````python
def buscar_pm25_brasil(limite=100):
    """
    Busca medições de PM2.5 no Brasil via API OpenAQ v2.
    Retorna lista de dicionários com chaves: 'valor', 'cidade', 'data'.
    """
    url = (
        f"https://api.openaq.org/v2/measurements"
        f"?country=BR&parameter=pm25&limit={limite}&sort=desc&order_by=datetime"
    )
    try:
        req = urllib.request.Request(url, headers={"Accept": "application/json"})
        with urllib.request.urlopen(req, timeout=10) as response:
            dados_brutos = json.loads(response.read().decode())
        registros = []
        for m in dados_brutos.get("results", []):
            registros.append({
                "valor": m["value"],
                "cidade": m.get("city", "Desconhecida"),
                "data":  m["date"]["local"][:10]
            })
        return registros
    except Exception as e:
        print(f"⚠️  Não foi possível acessar a API: {e}")
        return None


registros = buscar_pm25_brasil()

if registros:
    print(f"✅ {len(registros)} medições obtidas da API OpenAQ.")
    print("Primeiros 3 registros:", registros[:3])
else:
    print("Usando dados de fallback (próxima célula).")
````
